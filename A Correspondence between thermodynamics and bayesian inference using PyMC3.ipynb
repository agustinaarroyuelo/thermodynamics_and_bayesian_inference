## Introduction

On "Molecular Driving Forces" Ken A. Dill and Sarina Bromberg state that Statistical Thermodynamics provide a set of tools for modeling molecular forces and behaviour, as well as stablish a language for interpreting experiments. Statistical Thermodynamics as a discipline, tries to find wich are the forces that drive nature occurring processes and for this task, stablishes a bridge between measurable quantities (density, temperature, pressure, heat capacity, molecular radius, equilibrium constants) and the not measurable but underlying properties of energy, entropy, enthalpy and free energy. The latter ones gives us the posibility to predict equilibirum conditions in systems, and the former ones do not. 
A correspondence between statistical mechanics and bayesian inference has been discussed extensively. Most of the analogies between statiscal and thermodynamic concepts arise from the expression of marginal likelihood (bayesian evidence):

$$ p(\mathbb{X}|\alpha) = \int_{\theta}  p(\mathbb{X}|\theta)\  p(\theta|\alpha)\  d\theta$$ where $\theta \sim p(\theta|\alpha)$, a correspondence is established such that:

$$ Z(X^N) \equiv  \int_{\theta}\  e^{[-NĤ_X(\theta)]}\ \omega(\theta)\  d\theta$$ where $Ĥ_X(\theta) \equiv -log\ p(X|\theta)_{X \in x^N}$

Critical sample size $N_0$ is where the information content of the observations is equal to the information content of the prior. Computed as: $$N_0 \equiv \cfrac{\sigma^2}{\sigma^2_\omega}$$

```python
def N0(pop, sd):
    """
    Computes the critical sample size for a given population standard deviation.
       
    Parameters:
    pop: array. normal distribution
    
    Returns:
    N0: int. Critical sample size
    """
    return pop.var() / sd ** 2 
```

Relative sample size: $$\cfrac{N}{N_0}$$

```python
def relative_sample_size(pop, N0):
    """
    Compute relative sample size N/N0.
        
    Parameters:
    pop: array. normal distribution
    N0: int. Critical sample size
    
    Returns:
    rss: int. N/N0
    """
    return np.array([i / N0 for i in range(1, len(pop))])
```

Learning capacity $$\cfrac{K}{2\ \left(1+\cfrac{N_0}{N}\right)^2}$$

```python
def learning_capacity(pop, N0):
    """
    Compute Learning Capacity for a one-dimensional model. K=1.
        
    Parameters:
    pop: array. normal distribution
    N0: int. Critical sample size
    
    Returns:
    lc: array. Learning capacity 
    """
    r = [N0/i for i in range(1, len(pop))]
    lc = np.array([1 / (2 * (1 + i) ** 2) for i in r])
    return lc
```

Gibss entropy: $$ \dfrac{K}{2}\left(\cfrac{N}{N+N_0}-log\cfrac{N_0+N}{N_0}\right)$$

```python
def gibbs_entropy(pop, N0):
    """
    Compute Gibbs Entropy for a one-dimensional model. K=1.
        
    Parameters:
    pop: array. normal distribution
    N0: int. Critical sample size
    
    Returns:
    array. Gibbs Entropy 
    """
    a = np.array([i / i + N0 for i in range(1, len(pop))])
    b = np.array([- np.log(N0 + i / N0) for i in range(1, len(pop))])
    return np.add(a, b) / 2
```

```python
def plot_curves(plot, pop, sd):
    """
    Function to plot learning capacity and gibbs entropy curves
    """
    n0 = N0(pop, sd)
    rss = relative_sample_size(pop, n0)
    
    if plot == 'lc':
        lc = learning_capacity(pop, n0)
        plt.title('Learning Capacity', fontsize=18)
        plt.xlabel('Relative Sample Size $N/N_0$', fontsize=14)
        plt.ylabel('C', fontsize=14)
        plt.plot(rss, lc)
        plt.xscale('log')
        plt.axvline(x=n0/n0, linestyle='dashed', c='C4', label='Critical Sample Size')
        plt.legend(loc='best', fontsize=14)
        
    elif plot == 'ge':
        ge = gibbs_entropy(pop, n0)
        plt.title('Gibbs Entropy', fontsize=18)
        plt.xlabel('Relative Sample Size $N/N_0$', fontsize=14)
        plt.ylabel('S', fontsize=14)
        plt.plot(rss, ge)
        plt.axhline(y=ge.max(), linestyle='dashed', c='C2')
        plt.axvline(x=n0/n0, linestyle='dashed', c='C3')
        plt.xscale('log')
```

### Equipartition and learning capacity:
The central idea of the equipartition is that, in thermal equilibrium, the energy is distributed in a system in equal parts among its various dregrees of freedom; for example, the average kinetic energy in a movement of translation of a molecule must be equal to the average kinetic energy in its movement of rotation.

Equipartition predicts that the reduced heat capacity should be constant, equal to half the degrees of freedom in the system. 

At low sample size N (high temperature), the parameters are completely specified by model constraints (the prior) and therefore the parameters do not contribute to the learning capacity. At large sample size N, the parameters become data
dominated and therefore the learning capacity is predicted by equipartition ($\cfrac{1}{2}K$)

```python
plot_curves('ge', true1, sd)
```

### Gibss Entropy and model weighting
Since the Gibbs entropy is understood as the log-number of indistinguishable models at sample size N, constant entropy
is understood as equal model weighting.

While the Gibbs entropy remains constant, the weight of the models is uniform and the predictive capacity is low. As the number of observations N increases, the entropy decreases, the predictive capacity increases and the weight of the models is not uniform, one of the models has more and more weight.

From the previous examples it should be clear that as data samples increase, the results are less and less sensitive to the prior. In fact, for an infinite amount of data the prior has no effect. Exactly how much data is necessary for the a priori effect to be negligible varies according to the problem and the models used.

```python
# simulate true distribution, sd's and variances
true_data = np.random.normal(loc=10, scale=100, size=10000)
#true_data = np.random.uniform(low=10, high=100, size=10000)
standard_deviations = np.linspace(1, 1.5, 9)
variances = standard_deviations ** 2
```

```python
critical_ss = np.array([N0(true_data, i) for i in variances])
relative_ss = np.array([relative_sample_size(true_data, i) for i in critical_ss])
learning_capacities = np.array([learning_capacity(true_data, i) for i in critical_ss])
gibbs_entropies = np.array([gibbs_entropy(true_data, i) for i in critical_ss])
T = [1/i for i in range(1, 10000)]
```

```python
# plot learning capacity evolution
fig, ax = plt.subplots(3, 3, sharex=True, sharey=False)
ax = np.ravel(ax)
for i, lc in enumerate(learning_capacities):
    ax[i].plot(relative_ss[i], lc, linewidth=6, c='C2')
    ax[i].set_xlabel('$N_0$: {:.1f}'.format(critical_ss[i]), fontsize=10)
    ax[i].set_title('$\sigma^2_\omega:${:.2f}'.format(variances[i]), fontsize=10)
plt.tight_layout()
```

## Conclussions

* From the equation $ N = \dfrac{1}{T}$ we derive that when $N \rightarrow \infty$ then $T \rightarrow 0$. The system is completely defined. And when $N \rightarrow 0$ then  $T \rightarrow \infty$. The system has an infinit number of possible states.
* If $N_0 \equiv \cfrac{\sigma^2}{\sigma^2_\omega}$, then when $N_0 \rightarrow 0$ we are in the limit of the model with no prior information about the mean. And when $N_0 \rightarrow \infty$ is the limit where prior information completely specifies the mean. 
* When $N \rightarrow \infty$ and $T \rightarrow 0$, the learning capacity $C$ is predicted by equipartition as $\dfrac{1}{2}K$ where K is the number of dimensions for the model. When $N \rightarrow 0$ and $T \rightarrow \infty$, the model parameters $\theta$ are completely specified by the prior $\omega(\theta)$ and they do not contribute to the learning capacity.
* From Gibss Entropy analysis a condition for equal model weighting can be derived $\overline{S}\ (\theta_0; N, \omega) \approx const$.
* Uncertainty about a parameter value (hpd), decreases more rapidly around the $N_0$ sample size.
